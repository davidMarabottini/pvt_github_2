```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
skills = {
    'Frontend': 9,
    'Data': 8,
    'Scrum': 8,
    'SQL': 9,
    'Matematics': 8
}
```

```python
def print_radar_chart(skills):
  
  labels, values = [chiave for chiave in skills], [valore for valore in skills.values()]
  num_vars = len(labels)
  values += values[:1]

  angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
  angles += angles[:1]

  _, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))

  ax.plot(angles, values, linewidth=2, linestyle='solid', label='Livello di competenza')
  ax.fill(angles, values, 'b', alpha=0.25)

  ax.set_xticks(angles[:-1])
  ax.set_xticklabels(labels, fontsize=12, fontweight='bold')

  ax.set_ylim(0, 10)

  plt.show()
```

```python
print_radar_chart(skills)
```

```python
labels, values = [chiave for chiave in skills], [valore for valore in skills.values()]
plt.bar(labels, values)
plt.show()
```
